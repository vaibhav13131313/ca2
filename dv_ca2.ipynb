```python
nos=df['nose_long'].value_counts().reset_index()
nos
```

```python
sns.set_style("darkgrid")
colors = sns.color_palette("pastel")
plt.figure(figsize=(10, 6))
sns.barplot(x='index', y='nose_long', data=nos, palette=colors)
plt.xlabel('nose_long', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('nose_long Level Distribution', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

```python
long_hair=df['long_hair'].value_counts().reset_index()
long_hair
```

```python
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='index', y='long_hair', data=long_hair, palette='viridis')
plt.xlabel('long_hair', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('long_hair Distribution', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.tight_layout()
plt.show()
```

```python
lips_thin_counts = df['lips_thin'].value_counts()
# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(lips_thin_counts, labels=lips_thin_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of lips_thin')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
long_hair_counts = df['long_hair'].value_counts()
# Plot a pie chart for City
plt.figure(figsize=(8, 8))
color=['#706EEB','#5756B7','#B7B9EE']
plt.pie(long_hair_counts, labels=long_hair_counts.index, autopct='%1.1f%%', startangle=140,colors=color)
plt.title('Distribution of long_hair')
plt.axis('equal')
plt.show()
```

```python
cross_tab = pd.crosstab(df['forehead_width_cm'], df['forehead_height_cm'])
custom_colors = ["#D891EE", "#C649EB", "#9A39B8"]
plt.figure(figsize=(40, 10))
cross_tab.plot(kind='bar', stacked=True, color=custom_colors)
plt.xlabel('forehead_width_cm')
plt.ylabel('Count')
plt.title('Distribution of forehead_height_cm by forehead_width_cm')
plt.xticks(rotation=65)
plt.legend(title='forehead_height_cm', bbox_to_anchor=(1.00005, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.scatter(df['forehead_width_cm'], df['forehead_height_cm'])
plt.xlabel('forehead_width_cm')
plt.ylabel('forehead_height_cm')
plt.title('Scatter Plot of forehead_width_cm vs forehead_height_cm')
plt.show()
```

```python
cross_tab_city_leave = pd.crosstab(df['nose_wide'], df['nose_long'])
plt.figure(figsize=(10, 6))
cross_tab_city_leave.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.xlabel('nose_wide')
plt.ylabel('Count')
plt.title('Distribution of nose_wide by nose_wide')
plt.xticks(rotation=45)
plt.legend(title='nose_wide', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
numerical_columns = ['nose_wide', 'nose_long', 'lips_thin']
numerical_data = df[numerical_columns]
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df[['distance_nose_to_lip_long', 'long_hair','lips_thin']])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
attributes = [ 'forehead_width_cm', 'forehead_height_cm']
for attribute in attributes:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='gender', y=attribute, data=df)
    plt.title(f'Box Plot of {attribute} by Gender')
    plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='nose_wide', hue='forehead_height_cm', data=df, palette='Set2', dodge=True)
plt.title('Distribution ofnose wide by nose long')
plt.xlabel('nose_long')
plt.ylabel('Count')
plt.legend(title='Nose', loc='upper right')
plt.show()
```
